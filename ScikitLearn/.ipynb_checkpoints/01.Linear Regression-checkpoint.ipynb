# 1. Linear Regression
### 공부 시간에 따른 시험 점수

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values # 처음부터 마지막 컬럼 직전까지의 데이터 (독립 변수 - 원인)
y = dataset.iloc[:, -1].values # 마지막 컬럼 데이터 (종속 변수 - 결과)
```

```python
X, y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() # 객체 생성
reg.fit(X, y) # 학습 (모델 생성)
```

```python
y_pred = reg.predict(X) # X 에 대한 예측 값
y_pred
```

```python
plt.scatter(X, y, color='red') # 산정도
plt.plot(X, y_pred, color='blue') # 선 그래프
plt.title('Score by hours') # 제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
print('9시간 공부했을 때 예상 점수 : ', reg.predict([[9], [8], [7]])) # 9,8,7 시간일때 예측 값
```

```python
reg.coef_ # 기울기 (m)
```

```python
reg.intercept_ # y 절편 (b)
```

### 데이터 세트 분리

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset
```

```python
X = dataset.iloc[:, :-1].values # dataset.iloc[:, 0].values 라고 쓰지않는 이유는, 독립변수가 여러개일 때, 출력하기 위함
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # 훈련 80 : 테스트 20 으로 분리

# random_state=0 이라는 것이 없다면, X,y에서 임의의 값을 훈련, 테스트 값으로 저장하기 때문에 이 코드가 실행될 때마다 train, test 값이 달라짐
```

```python
X_train, len(X_train)
```

```python
X_test, len(X_test)
```

### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train) # 훈련세트로 학습
```

```python
plt.scatter(X_train, y_train, color='red') # 산정도
plt.plot(X_train, reg.predict(X_train), color='blue') # 선 그래프
plt.title('Score by hours') # 제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test, y_test, color='red') # 산정도
plt.plot(X_test, reg.predict(X_test), color='blue') # 선 그래프
plt.title('Score by hours') # 제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

### 모델 평가

```python
reg.score(X_test, y_test) # 테스트 세트를 통한 모델 평가
```

```python
reg.score(X_train, y_train)
```

### 경사 하강법 (Gradient Descent)

max_iter = 훈련 세트 반복 횟수 (Epoch 횟수)

eta0 = 학습률 (learning rate)

```python
from sklearn.linear_model import SGDRegressor

# sr = SGDRegressor(max_iter=100, eta0=1e-4, random_state=0, verbose=1)
sr = SGDRegressor()

sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='red') # 산정도
plt.plot(X_train,  sr.predict(X_train), color='blue') # 선 그래프
plt.title('Score by hours (train data, SGD)') # 제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
sr.coef_, sr.intercept_
```

```python
sr.score(X_test, y_test)
```

```python
sr.score(X_train, y_train)
```
